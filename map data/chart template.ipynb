```python
import numpy as np

import plotly.offline as offline
offline.init_notebook_mode(connected=True)
import plotly.graph_objs as go
```

```python
scatter = np.loadtxt('map.csv', delimiter=',')
interp = np.loadtxt('interp.csv', delimiter=',')
xgrid = np.loadtxt('xgrid.csv', delimiter=',')
ygrid = np.loadtxt('ygrid.csv', delimiter=',')
```

```python
# check kriging result

trace1 = go.Scatter3d(
    x = scatter[:, 0],
    y = scatter[:, 1],
    z = scatter[:, 2],
    mode = 'markers',
    marker=dict(
        color='orange'
    )
)

trace2 = go.Surface(
    x = xgrid,
    y = ygrid,
    z = interp,
    opacity=0.95,
    colorscale='Volcano'
)

traces = [trace1, trace2]

camera = dict(
    up=dict(x=0, y=0, z=1),
    center=dict(x=0, y=0, z=0),
    eye=dict(x=-1.25, y=-1.25, z=1.5)
)

layout = go.Layout(
    width=800,
    height=700,
    autosize=False,
    scene=dict(
        xaxis=dict(
            title='x = bin index',
            gridcolor='rgb(255, 255, 255)',
            zerolinecolor='rgb(255, 255, 255)',
            showbackground=True,
            backgroundcolor='rgb(230, 230,230)'
        ),
        yaxis=dict(
            title='y = loop',
            gridcolor='rgb(255, 255, 255)',
            zerolinecolor='rgb(255, 255, 255)',
            showbackground=True,
            backgroundcolor='rgb(230, 230,230)'
        ),
        zaxis=dict(
            title='z = impact',
            gridcolor='rgb(255, 255, 255)',
            zerolinecolor='rgb(255, 255, 255)',
            showbackground=True,
            backgroundcolor='rgb(230, 230,230)'
        ),
        aspectratio = dict( x=1, y=1, z=0.7 ),
        aspectmode = 'manual',
        camera=camera
    )
)

fig = dict(data=traces, layout=layout)

offline.iplot(fig)
```
